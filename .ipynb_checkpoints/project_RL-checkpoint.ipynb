# 코드 설명

---
**코드의 도입부**

```python
%matplotlib inline

import gym
import collections
import random

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

---
**본 코드의 Hyperparameters들**

```python
#Hyperparameters
learning_rate = 0.001
gamma         = 0.99
buffer_limit  = 50000
batch_size    = 128

func_epsilon = lambda n_epi : max(0.001, 0.08 - 0.001*(n_epi/200))

layer = [6, 64, 64, 3]                       # 첫번째는 무조건 6(state 갯수), 마지막은 무조건 3(action 갯수)

print_interval = 20
plot_interval = 100
```

**learning_rate**는 Deep Network 학습 시, 학습률을 뜻하고

**gamma**는 q함수에서의 감가율을 뜻하고

**buffer_limit**은 Replay Buffer의 크기를,

**batch_size**는 학습 시, 사용할 Episode 갯수를 의미하며

**layer**는 Deep Network의 구조를 정의한다.

---
**DQN에서의 Replay를 수행하기 위한 Buffer클래스**

```python
class ReplayBuffer():
    def __init__(self):
        self.buffer = collections.deque(maxlen=buffer_limit)
    
    def put(self, data):
        self.buffer.append(data)
    
    def sample(self, n):
        mini_batch = random.sample(self.buffer, n)
        s_lst, a_lst, r_lst, s_prime_lst, done_mask_lst = [], [], [], [], []
        
        for transition in mini_batch:
            s, a, r, s_prime, done_mask = transition
            s_lst.append(s)
            a_lst.append([a])
            r_lst.append([r])
            s_prime_lst.append(s_prime)
            done_mask_lst.append([done_mask])

        return torch.tensor(s_lst, dtype=torch.float), torch.tensor(a_lst), \
               torch.tensor(r_lst), torch.tensor(s_prime_lst, dtype=torch.float), \
               torch.tensor(done_mask_lst)
    
    def size(self):
        return len(self.buffer)
```

**ReplayBuffer**는 DQN에서 Replay를 실현시키기 위해서 만들어진 클래스로서 4개의 메서드를 가지고 있다.
- **생성메서드** : 객체 생성 시, 최대 크기가 buffer_limit인 deque를 만든다.
- **put 메서드** : 실행 시, 객체의 buffer변수에 들어온 매개변수 data를 추가시킨다.
- **sample 메서드** : 실행 시, 객체의 buffer변수 값 중에서 매개변수 n개의 Episode를 꺼내 s, a, r, s_prime, done_mask로 분리 후 각각을 모은다. 모아놓은 배열들을 torch.tensor로 wrapping 시킨 후 return한다.
- **size 메서드** : 객체의 buffer변수의 크기를 return한다.

---
**Deep Network 구현**

```python
class Qnet(nn.Module):
    def __init__(self):
        super(Qnet, self).__init__()
        self.fc1 = nn.Linear(layer[0], layer[1])
        self.fc2 = nn.Linear(layer[1], layer[2])
        self.fc3 = nn.Linear(layer[2], layer[3])
        

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
      
    def sample_action(self, obs, epsilon):
        out = self.forward(obs)
        coin = random.random()
        if coin < epsilon:
            return random.randint(0,2)
        else : 
            return out.argmax().item()
```

**Qnet**는 DQN에서 Deep Network를 실현시키기 위해서 만들어진 클래스로서 3개의 메서드를 가지고 있다. 해당 클래스는 torch.nn.Module을 상속받고 있다.
- **생성메서드** : 객체 생성 시, 부모 클래스의 생성 메서드를 실행시키고 3층의 Ann의 신경망을 정의한다.
- **forward 메서드** : 실행 시, 각각의 Layer들이 relu로 연결된 신경망의 forward를 실행시킨 후, 그 값을 return한다.
- **sample_action 메서드** : 본 메서드는 epsilon-greedy를 구현한 메서드이다. 실행 시, epsilon의 확률로 랜덤한 Action을 return하고 그 외의 확률로 학습된 정책에 의거한 Action을 return한다.

---
**학습을 위한 메소드**

```python
def train(q, q_target, memory, optimizer):
    for i in range(10):
        s,a,r,s_prime,done_mask = memory.sample(batch_size)

        q_out = q(s)
        q_a = q_out.gather(1,a)
        max_q_prime = q_target(s_prime).max(1)[0].unsqueeze(1)
        target = r + gamma * max_q_prime * done_mask
        loss = F.smooth_l1_loss(q_a, target)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
```

함수 **train**는 DQN에서 오류함수를 이용하여 Deep Network를 학습시키는 함수이다. 오류함수는 다음과 같다.

![](img/dqn_errer_equation.png)
여기서 참값은 fixed Q-targets을 이용한다.

---
**코드 구동을 위한 main함수**

```python
def main():
    
    epi_list = []
    score_list = []
    
    env = gym.make('Acrobot-v1')
    q = Qnet()
    q_target = Qnet()
    q_target.load_state_dict(q.state_dict())                            # 주기적으로 q_target을 업데이트한다.
    memory = ReplayBuffer()
    optimizer = optim.Adam(q.parameters(), lr=learning_rate)
    
    score = 0.0  
    render = False
    episode_in_sample = 0
    
    for n_epi in range(10000):
        
        episode_in_sample += 1
        
        epsilon = func_epsilon(n_epi)                                   #Linear annealing from 8% to 0.1%
        s = env.reset()
        done = False

        while not done:
            a = q.sample_action(torch.from_numpy(s).float(), epsilon)   # epsilon-greedy에 따라 0 또는 1을 취한다.       
            s_prime, r, done, info = env.step(a)                        # 특정 action에 대한 환경의 정보를 출력한다.
            done_mask = 0.0 if done else 1.0                           # 게임이 끝났다면 0, 끝나지 않았다면 1
            memory.put((s,a,r,s_prime, done_mask))
            s = s_prime

            score += r
            
            if render:
                env.render()
                
            if done:
                break
            
        if (n_epi > 5000) and (SCORE_for_render > -100):
            render = True
            
        if memory.size() > 2000:
            train(q, q_target, memory, optimizer)

        if (n_epi % print_interval==0) and (n_epi != 0):
            q_target.load_state_dict(q.state_dict())                   # Print interval마다 Q_target을 업데이트해준다.
            print("n_episode :{}, score : {:.1f}, n_buffer : {}, eps : {:.1f}%".format(
                                                            n_epi, score/print_interval, memory.size(), epsilon*100))
            epi_list.append(n_epi)
            score_list.append(score / print_interval)
            SCORE_for_render=score/print_interval
            episode_in_sample = 0
            score = 0.0

        if n_epi % plot_interval == 0 and n_epi != 0:
            
            fig = plt.figure()
            plt.scatter(epi_list, score_list)
            plt.title("epi : {}".format(n_epi))
            plt.xlabel('episode')
            plt.ylabel('score')
            plt.show()
            
            data = pd.DataFrame({'episode' :epi_list , 'score' : score_list})
            data.to_excel('data.xlsx')
    env.close()
    
    return epi_list, score_list

if __name__ == '__main__':
    epi_list, score_list = main()
    
    fig = plt.figure()
    plt.scatter(epi_list, score_list)
    plt.title("final graph")
    plt.xlabel('episode')
    plt.ylabel('score')
    plt.show()
```

# 학습의 어려움

다른 코드들을 참고해보아도 학습률이 중간에 튀는 경우가 많음을 알 수 있다.

이에 대한 원인을 인터넷 등으로 찾아보았을 때, 이중 진자는 단일 진자에 비해 다소 분석하기도 어렵다는 것이 

알려져 있다. 이중 진자의 경우 그림처럼 단진자보다 훨씬 더 복잡한 움직임을 보인다. 이중 진자의 움직임을 

나타낸 수식을 보면 알 수 있듯이, 이중 진자의 움직임을 나타내기 위해서 복잡한 식이 필요하기 때문에 학습하는 데
